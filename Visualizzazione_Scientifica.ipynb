# Di che parliamo?

### 1) Quale costruttore ha più vittorie?

##### Apriamo file dei results

```python
results = pd.read_csv('Dataset/results.csv')
results.head()
```

```python
wins = results[results['positionOrder'] == 1].merge(constructors, on='constructorId')
wins
```

```python
wins['name'].value_counts().plot.bar()
```

### 2) Quando la Ferrari (scuderia che ha più vittorie) ha vinto negli anni?

```python
fer_wins = results[(results['positionOrder'] == 1) & (results['constructorId'] == 6)].merge(races, on = 'raceId')
fer_wins.columns
# fer_wins[['resultId', 'raceId', 'driverId', 'position', 'points', 'year', 'round', 'name']]
fer_wins['year'].value_counts().sort_index().plot.bar()
```
